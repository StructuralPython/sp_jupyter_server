# Calculate Concrete Beam Stirrup Spacing

Assumes:
* V_f is already reduced by a distance d_v from face of support
* Simplified method can be used
* No torsion

```python
## Inputs

# Loads:
V_f = 204*kip

# Beam Geometry:
b_w = 6*ft
d = 3*ft - 4*inch
h = 3*ft
f_prime_c = 35*MPa
f_y = 400*MPa
phi_c = 0.65
phi_s = 0.85

# Assumed Stirrups Area
A_v = 400* mm**2
```

## Loads

```python
%%render params
V_f
```

## Beam Geometry

```python
%%render params
b_w
h
d
f_prime_c
phi_c
phi_s
A_v
```

## Calculate $d_v$

```python
%%render
d_v = min(0.9*d, 0.72*h)
```

## Calculate $\beta$

```python
%%render
beta = (230*mm) / (1000*mm + d_v)
```

```python
import math

def sqrt(x):
    """
    This is a fancy sqrt() function that automatically detects if it is 
    operating on an MPa unit and adjusts accordingly.
    """
    if hasattr(x, "dimensions") and x.dimensions == (1, -1, -2, 0, 0, 0, 0):
        return math.sqrt(x)*MPa
    else:
        return math.sqrt(x)
```

```python
%%render long
V_c = phi_c * beta * sqrt(f_prime_c) * b_w * d_v
```

## Evaluate $V_f$ against $V_c$

```python
%%render
if h > 750*mm: V_s = V_f - V_c
elif V_f >= V_c: V_s = V_f - V_c
elif V_f < V_c: V_s = 0*kN
```

## Calculate $s_{max}$

```python
%%render
s_max_1 = min(0.7*d_v, 600*mm)
s_max_2 = (A_v * f_y) / (0.06 * b_w * sqrt(f_prime_c))
s_max = min(s_max_1, s_max_2)
```
